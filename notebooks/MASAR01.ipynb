```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core.storage.storage_context import StorageContext
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# 1️⃣ Load the PDF
documents = SimpleDirectoryReader(input_files=["/content/TBS_Handbook-2022.pdf"]).load_data()

# 2️⃣ Create index
# Use a local embedding model since no OpenAI API key is provided
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)

# 3️⃣ Save (persist) index to folder
index.storage_context.persist(persist_dir="./pdf_index")

print("✅ LlamaIndex created and saved to ./pdf_index")
```
